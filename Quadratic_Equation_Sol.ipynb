```python
from time import sleep
import math
import cmath

a = int(input("a from ax^2+bx+c = 0:"))
b = int(input("b from ax^2+bx+c = 0:"))
c = int(input("c from ax^2+bx+c = 0:"))

put = (b**2)-(4*a*c)
if put >= 0:
  sqrt = -1*b + math.sqrt(put)
if put < 0:
  sqrt = -1*b + cmath.sqrt(put)
x1 = sqrt/(2*a)
x2 = -1*(b/a + x1)

print("I'm now calculating...")
sleep(3)
print("Yes! I found the answer!")
sleep(2)
print("You'd like to know it, right?")
sleep(2)
reaction = input("Select [Yes] [No]:")

if reaction=='Yes':
  print('The answer is...')
  sleep(1)
  if x1!=x2:
    print(x1,",",x2)
  else:
    print(x1)

while reaction!='Yes':
  print("Oops!...")
  sleep(1)
  reaction = input("Select [Yes] [No]:")
  if reaction=='Yes':
    print("Answer:",x1,",",x2)
    break
```
